```python
from chroma import Chroma
from embeddings import OpenAIEmbeddings
import os
import chromadb
import json
from tqdm import tqdm
os.environ['OPENAI_API_KEY'] = ''
import pickle
```

```python
embd = OpenAIEmbeddings()
client = chromadb.PersistentClient(path="chroma1000")
db = Chroma(client=client, embedding_function=embd, persist_directory='chroma1000')
```

```python
case5000 = os.listdir('case1000')
case5000 = [i for i in case5000 if i.endswith('.json')]
data = dict()
for i in range(len(case5000)):
    with open('case1000/' + case5000[i], 'r') as file:
        try:
            d = json.load(file)
            data[case5000[i]] = d
            case5000[i] = d
        except:
            print('error in file: case5000/' + case5000[i])
for i in case5000:
    assert 'casebody' in i
    assert 'data' in i['casebody']
# pickle.dump(data, open('case1000_name2data.pkl', 'wb'))
```

```python
def get_text(data):
    if isinstance(data, str):
        return data
    if isinstance(data, list):
        return '   \n'.join([get_text(i) for i in data])
    if isinstance(data, dict):
        return get_text(data['text'])
```

```python
data = []
metadata = []
for id, i in enumerate(case5000):
    metadata.append(i)
    data.append("")
    for k, v in i['casebody']['data'].items():
        if isinstance(v, str):
            data[id] += f"{k}: {v}\n"
        elif isinstance(v, list):
            tmp = "   \n".join(map(get_text, v))
            data[id] += f"{k}: \n{tmp}\n"
        else:
            print(f"Unknown type: {type(v)}")
    
```

```python
HASH = dict()
for i in range(len(data)):
    HASH[data[i]] = metadata[i]
pickle.dump(HASH, open('case1000_content2metadata.pkl', 'wb')) 
```
